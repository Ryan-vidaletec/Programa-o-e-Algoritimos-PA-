```python
# Operadores aritméticos

# Fazer um programa que o usuário digita 02 números
# Saída mostrar operações aritméticas.

# Entrada

a = int(input("digite um número: "))
b = int(input("digite um número:"))

# Processamento

soma = a + b
subtração = a - b
multiplicação = a * b
divisão = a / b
div_int = a // b
resto = a % b
potencia = a ** b

# Saída

print("")
print("------------------------OPERADORES ARITMÉTICOS------------------------")
print("                                                                      |")
print("------------------------------SOMA-------------------------------------")
print("                                                                      |")
print(f"O resultado da SOMA de {a} mais {b} é igual a: {soma}")
print("                                                                      |")
print("---------------------------SUBTRAÇÃO----------------------------------")
print("                                                                      |")
print(f"O resultado da SUBTRAÇÃO de {a} mais {b} é igual a: {subtração}")
print("                                                                      |")
print("--------------------------MULTIPLICAÇÃO--------------------------------")
print("                                                                      |")
print(f"O resultado da MULTIPLICAÇÃO de {a} mais {b} é igual a: {multiplicação}")
print("                                                                      |")
print("-----------------------------DIVISÃO-----------------------------------")
print("                                                                      |")
print(f"O resultado da DIVISÃO de {a} mais {b} é igual a: {divisão}")
print("                                                                      |")
print("-----------------------------DIVISÃO-----------------------------------")
print("                                                                      |")
print(f"O resultado da DIVISÃO de {a} mais {b} é igual a: {div_int}")
print("                                                                      |")
print("------------------------------RESTO------------------------------------")
print("                                                                      |")
print(f"O resultado do RESTO de {a} mais {b} é igual a: {resto}")
print("                                                                      |")
print("-----------------------------POTÊNCIA----------------------------------")
print("                                                                      |")
print(f"O resultado da POTÊNCIA de {a} mais {b} é igual a: {potencia}")
```

```python
# Fazer um programa que converte horas em minutos e horas em segundos

# Entrada
horas = float(input("Digite o número de horas: "))

# Processamento
minutos = horas * 60
segundos = horas * 3600

# Saída
print(f"{horas} horas equivalem a {minutos} minutos.")
print(f"{horas} horas equivalem a {segundos} segundos.")
```

```python
# Fazer um programa que converte km em metros

# Entrada
km = float(input("Digite a distância em KM: "))

# Processamento
metros = km * 1000

# Saída
print(f"{km} KM equivale a {metros} metros.")
```

```python
# Fazer um programa que converte km em milhas

# Entrada
km = float(input("Digite a distância em KM:"))

# Processamento
milhas = km / 1.609

# Saída
print(f"{km} KM equivalem a {milhas} milhas")
```
