```python
# Calcula área do retângulo

# Entrada

comprimento = float (input("Digite o valor do comprimento:"))
altura = float (input("Digite o valor da altura:"))


# Processamento

área = comprimento * altura


# Saída


print(f"A área do retângulo é igual: {área}")
```

```python
# Escreve um programa que calcula o dobro de um número

# Entrada

número = int (input("Digite um número:"))

# Processamento

dobro = número * 2

# Saída

print (f"O dobro de {número} é igual a: {dobro}")
```
